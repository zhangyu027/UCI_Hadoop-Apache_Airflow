### **Step 1: Installing Hadoop**

Hadoop is an open-source, Java-based framework designed to process and store large volumes of data across a distributed network of inexpensive machines. It is a key tool in the big data ecosystem, providing the infrastructure needed to handle massive datasets efficiently. Initially developed by the Apache Software Foundation, Hadoop was the first significant open-source project in the field of big data.

In this step, you will learn how to install Hadoop on your machine or cluster. This will allow you to take full advantage of its distributed storage system (HDFS) and data processing capabilities (MapReduce) for large-scale data analysis.

### **Step 2: Hadoop WordCount Example**

The Grep example searches for a pattern. The WordCount example counts how many times each word appears in the input files.

In this section, we will:

1. Create a small text file.
2. Put the text file into the Hadoop input folder.
3. Run Hadoop WordCount.
4. View the WordCount output.

**Instructor note:** The solution cells below can be hidden before sharing the notebook with students so they can practice the commands themselves.

````
# Create a new text file for the WordCount example
# This file will be added to the existing input folder.

%%bash
cat > ~/input/wordcount_sample.txt << 'EOF'
Hadoop helps process large data
Hadoop MapReduce counts words
Big data tools include Hadoop and Spark
Spark is often used with Hadoop
WordCount is a classic Hadoop example
EOF

echo "Created file:"
ls -l ~/input/wordcount_sample.txt

echo "File content:"
cat ~/input/wordcount_sample.txt
````

````
# Remove old WordCount output folder if it already exists.
# Hadoop requires the output folder to NOT exist before the job runs.

!rm -rf ~/wordcount_output
````

````
# Run Hadoop WordCount on the input folder.
# This reads inside ~/input,  the new wordcount_sample.txt file.

# Run Hadoop WordCount on only one file
!/content/hadoop-3.4.1/bin/hadoop jar \
/content/hadoop-3.4.1/share/hadoop/mapreduce/hadoop-mapreduce-examples-3.4.1.jar \
wordcount ~/input/wordcount_sample.txt ~/wordcount_output
````

````
# View the WordCount result.

!cat ~/wordcount_output/*
````
