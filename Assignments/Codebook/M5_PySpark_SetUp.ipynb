```
import os
from pyspark.sql import SparkSession

endpoint = os.environ.get("SPARK_CONNECT", "").strip()

if endpoint:
    # Connect to your Dataproc Spark Connect endpoint
    # Example (placeholder): sc://<region>-dataproc.googleapis.com:443/projects/<PROJECT>/locations/<REGION>/clusters/<CLUSTER>
    spark = SparkSession.builder.remote(endpoint).getOrCreate()
    print("Connected via Spark Connect ->", endpoint)
else:
    # Fallback to local session for development/testing
    spark = SparkSession.builder.master("local[*]").appName("RDDPractice").getOrCreate()
    print("Started local SparkSession (fallback). Set SPARK_CONNECT to use Dataproc Spark Connect.")
```
