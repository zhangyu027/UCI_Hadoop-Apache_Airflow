## 1. What is Customer Retention Rate?

Customer retention refers to the activities and strategies that businesses use to encourage customers to continue purchasing their products or services. This concept focuses on increasing the likelihood of repeat purchases, which ultimately boosts profitability. The reason why customer retention is so valuable is that each repeat purchase typically costs less in marketing and acquisition efforts than attracting new customers. Retention strategies can be both short-term or long-term, depending on the business's goals, but they both focus on building customer loyalty.

Customer retention should be a top priority for businesses because even a small improvement in retention—just 5%—can increase profits significantly, by anywhere between 25% to 95%. Just like customer acquisition, customer retention is crucial for growth and profitability. While retention may take longer to show results compared to acquiring new customers, in the long run, your business will rely more on retaining existing customers than constantly finding new ones.

Churn is the opposite of retention and refers to customers who stop returning to your business after their initial purchase. The churn rate is a key metric used to measure customer retention and to understand how well a business is performing at keeping its customers.

To improve retention, businesses can employ a variety of customer engagement strategies. By using data, businesses can identify customers who are more likely to stay and continue purchasing, while also spotting weak areas in the customer journey that could be improved.

## 2. What Is Customer Retention Analytics?

Customer retention analytics refers to the measurement of specific metrics that indicate how well a business is retaining its customers. These metrics help you understand how satisfied customers are, whether they are likely to stay loyal, and if they are on their way to becoming repeat buyers.

By analyzing customer retention data, businesses can gain valuable insights into how to improve their retention strategies. For example, if a business finds that 80% of its revenue comes from just 20% of its customer base, it can use retention analytics to focus on keeping that critical group of customers.

Data-driven insights can also be used to predict customer behavior using techniques such as machine learning. By analyzing data like purchase history, demographics, and customer preferences, businesses can anticipate which customers might be at risk of leaving and take action to retain them.

## How Does Data Analytics Benefit Customer Retention?

- **Minimizes Customer Acquisition Costs (CAC)**: Retaining existing customers is typically less expensive than acquiring new ones.
- **Creates Sustainable Growth**: Focusing on retention can lead to steady, long-term growth without relying solely on new customer acquisition.
- **Provides Quality Insights Into Customer Experience**: Data helps businesses understand their customers' needs, preferences, and pain points, enabling them to improve the overall customer experience.
- **Establishes Customer Loyalty**: With the right insights and strategies, businesses can create stronger relationships with customers, fostering loyalty and repeat business.

## Strategies to Reduce Churn

- **Prioritize High-Quality Leads**: Focus on acquiring and retaining high-value customers who are more likely to remain loyal.
- **Develop a Data Roadmap**: Organize and prioritize the data that will help in understanding and improving retention efforts.
- **Determine Machine Learning Methods**: Use machine learning algorithms to predict customer behavior and identify at-risk customers.
- **Set Key Performance Indicators (KPIs)**: Establish clear KPIs to track the effectiveness of your retention efforts and ensure that strategies are working.

Segment your customers based on specific characteristics, such as demographics or buying behaviors, to understand which groups are more likely to stay loyal. This allows for more targeted retention efforts that can improve customer satisfaction and reduce churn.

## A.) Demographics - Let us first understand the gender, age range, patner and dependent status of the customers

The data suggests that there is a positive correlation between churn and factors such as month-to-month contracts, lack of online security, and the absence of tech support. In other words, customers who are on month-to-month contracts or do not have access to online security and tech support are more likely to churn.

On the other hand, longer tenure and two-year contracts appear to be negatively correlated with churn, meaning that customers with longer contract durations are less likely to leave.

Interestingly, services like online security, streaming TV, online backup, and tech support—when offered without an internet connection—seem to have a negative relationship with churn. This suggests that customers who subscribe to these services are less likely to cancel, even if they do not have an internet connection.

After going through the above EDA we will develop some predictive models and compare them.

```python
## 1. Logistic Regression
# We will use the data frame where we had created dummy variables
y = df_dummies['Churn'].values
X = df_dummies.drop(columns = ['Churn'])

# Scaling all the variables to a range of 0 to 1
from sklearn.preprocessing import MinMaxScaler
features = X.columns.values
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(X)
X = pd.DataFrame(scaler.transform(X))
X.columns = features
```

```python
# Create Train & Test Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Running logistic regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
result = model.fit(X_train, y_train)
```

```python
from sklearn import metrics
prediction_test = model.predict(X_test)
# Print the prediction accuracy
print (metrics.accuracy_score(y_test, prediction_test))
```

```python
# To get the weights of all the variables
weights = pd.Series(model.coef_[0],
                 index=X.columns.values)
print (weights.sort_values(ascending = False)[:10].plot(kind='bar'))
```

```python
print(weights.sort_values(ascending = False)[-10:].plot(kind='bar'))
```

## Observations

Upon analyzing the data, we can identify several variables that either increase or decrease the likelihood of customer churn. A negative relationship with churn indicates that the likelihood of churn decreases when that particular variable is present. Below, we summarize some of the key findings:

- **Two-Month Contracts and Tenure:** As observed in our exploratory data analysis (EDA), having a two-month contract significantly reduces the likelihood of churn. Additionally, both the two-month contract and longer tenure have the strongest negative relationships with churn, as confirmed by the logistic regression model.
  
- **DSL Internet Service:** Customers with DSL internet service are less likely to churn. This suggests that DSL service may contribute to improved customer retention.

- **Factors Contributing to Higher Churn Rates:** On the flip side, variables such as total charges, month-to-month contracts, fiber optic internet services, and seniority seem to increase the likelihood of churn. It is particularly interesting to note that despite fiber optic services being faster and generally considered a premium offering, customers with fiber optic internet are more likely to churn. This warrants further exploration to understand why this counterintuitive trend exists.

In the following analysis, we will delve deeper into these variables to gain a better understanding of the factors driving customer churn.

### 2). Random Forest

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=101
)

# ✅ Fixed RandomForest parameters
model_rf = RandomForestClassifier(
    n_estimators=1000,
    oob_score=True,
    n_jobs=-1,
    random_state=50,
    max_features="sqrt",      # ← "auto" deprecated; use "sqrt" for classification
    max_leaf_nodes=30
)

# Train model
model_rf.fit(X_train, y_train)

# Evaluate
prediction_test = model_rf.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test, prediction_test))
```

```python
importances = model_rf.feature_importances_
weights = pd.Series(importances,
                 index=X.columns.values)
weights.sort_values()[-10:].plot(kind = 'barh')
```

## Observations:

From random forest algorithm, monthly contract, tenure and total charges are the most important predictor variables to predict churn.
The results from random forest are very similar to that of the logistic regression and in line to what we had expected from our EDA
