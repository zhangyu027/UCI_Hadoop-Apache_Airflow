###**Step 7.1:** Determine the patents that have been cited more than 10 times

Create Script 🐖

```python
# Create modified Pig script
%%writefile patent_citations.pig
/* patent_citations.pig */

-- Load the patent citation data
citations = LOAD 'file:///content/data_pig/cite75_99.txt' USING PigStorage(',')
    AS (citing:chararray, cited:chararray);

-- Filter out the header row based on the fact that header contains non-numeric values
citations_filtered = FILTER citations BY (citing matches '\\d+' AND cited matches '\\d+');

-- Convert the citing and cited columns to integers after filtering the header
citations_int = FOREACH citations_filtered GENERATE (int)citing AS citing, (int)cited AS cited;

-- Group by cited patent to count citations for each patent
grouped_citations = GROUP citations_int BY cited;

-- Count the number of times each patent has been cited
citation_counts = FOREACH grouped_citations GENERATE group AS cited, COUNT(citations_int) AS citation_count;

-- Filter to find patents cited more than 10 times
frequent_citations = FILTER citation_counts BY citation_count > 10;

-- Display the results
DUMP frequent_citations;
```

Run Script

```python
!pig -x local patent_citations.pig > results.txt
```

Inspect Results

```python
!head results.txt
!tail results.txt
```
